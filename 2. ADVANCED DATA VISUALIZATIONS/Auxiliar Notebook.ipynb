## Data transformation

```python
# Revision de columnas
data_raw.columns
```

```python
# Transformacion de dataframe
# Eliminando columnas innecesarias
cols_to_drop = ['unnamed:_66']
data_clean = data_raw.drop(columns=cols_to_drop, errors='ignore')

# Reestruturar dataframe
data_transformed = data_clean.melt(
    id_vars=['country_name', 'country_code', 'indicator_name', 'indicator_code'],
    var_name='year',
    value_name='value'
)

# Revision
print('-'*200)
print(f'Shape: {data_transformed.shape}')
print('-'*200)
print('Head')
display(data_transformed.head(5))
print('-'*200)
print('Tail')
display(data_transformed.tail(5))
print('-'*200)
```

```python
# Revision de numero de indicadores
data_transformed['indicator_name'].nunique()
```

```python
# Revision de numero de anios
data_transformed['year'].nunique()
```

```python
# Quitando zonas geograficas
non_countries = ['Arab World', 'Caribbean small states',
       'Central Europe and the Baltics', 'Early-demographic dividend',
       'East Asia & Pacific',
       'East Asia & Pacific (excluding high income)',
       'East Asia & Pacific (IDA & IBRD)', 'Euro area',
       'Europe & Central Asia',
       'Europe & Central Asia (excluding high income)',
       'Europe & Central Asia (IDA & IBRD)', 'European Union',
       'Fragile and conflict affected situations',
       'Heavily indebted poor countries (HIPC)', 'High income',
       'IBRD only', 'IDA & IBRD total', 'IDA blend', 'IDA only',
       'IDA total', 'Late-demographic dividend',
       'Latin America & Caribbean',
       'Latin America & Caribbean (excluding high income)',
       'Latin America & Caribbean (IDA & IBRD)',
       'Least developed countries: UN classification',
       'Low & middle income', 'Low income', 'Lower middle income',
       'Middle East & North Africa',
       'Middle East & North Africa (excluding high income)',
       'Middle East & North Africa (IDA & IBRD)', 'Middle income',
       'North America', 'OECD members', 'Other small states',
       'Pacific island small states', 'Post-demographic dividend',
       'Pre-demographic dividend', 'Small states', 'South Asia',
       'South Asia (IDA & IBRD)', 'Sub-Saharan Africa',
       'Sub-Saharan Africa (excluding high income)',
       'Sub-Saharan Africa (IDA & IBRD)', 'Upper middle income', 'World',
      ]

# Solo paises
data_countries_only = data_transformed[~data_transformed['country_name'].isin(non_countries)].reset_index(drop=True)

# Vista rpida
print(f"Shape: {data_countries_only.shape}")
print(f"Numero de paises:\n{data_countries_only['country_name'].nunique()}")
```
