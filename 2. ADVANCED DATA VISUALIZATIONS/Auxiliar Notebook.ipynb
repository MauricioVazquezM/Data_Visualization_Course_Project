```python
# Quitando indicadores con muchos missings
# Calcular cantidad de nulos por indicador
nulos_por_indicador = data_countries_only.groupby('indicator_name').agg(total_valores=('value', 'size'), 
                                                                        valores_nulos=('value', lambda x: x.isna().sum())).reset_index()

# Añadir columna de completitud (proporción de datos no nulos)
nulos_por_indicador['completitud'] = 1 - nulos_por_indicador['valores_nulos'] / nulos_por_indicador['total_valores']

# Ordenar por menor cantidad de nulos
top15_indicadores = nulos_por_indicador.sort_values(by='valores_nulos').head(10)
print(f'Cantidad de nulos en conjunto: {top15_indicadores.valores_nulos.sum()}')
top15_indicadores
```

```python
# Quedandonos con los indicadores con menos missings
indicadores_top15 = top15_indicadores['indicator_name'].tolist()
data_top15 = data_countries_only[data_countries_only['indicator_name'].isin(indicadores_top15)].reset_index(drop=True)
data_top15.indicator_name.unique()
```
